```python
#Importando bibliotecas

import numpy as np
from faker import  Faker
import matplotlib.pyplot as plt
import random
from wordcloud import WordCloud
```

```python
#Gerando dados, 10 linhas de nomes aleatórios e 10 linhas de pontuacoes aleatórias
#Gravando os dados em um arquivo

fake = Faker('pt_BR')
random.seed()
novalinha = '\n'

with open('nomesevalores.txt', 'a') as file:
    for _ in range(10):
        file.writelines(fake.name())
        file.writelines(novalinha)
    for _ in range(10):
        file.writelines(str(random.randrange(1, 10)))
        file.writelines(novalinha)
```

```python
#Abrindo os dados do arquivo gerado e criando listas com os dados

with open('nomesevalores.txt') as arquivo:
    linhas = []
    for linha in arquivo:
        linhas.append(linha.rstrip())

lista_nomes = linhas[0:10]
lista_pontuacao = linhas[10:20]
lista_pontuacao_int = [int(x) for x in lista_pontuacao]
pontuacoes = np.array(lista_pontuacao_int)

print(lista_nomes)
print(lista_pontuacao)
print(lista_pontuacao_int)
print(linhas)
print(pontuacoes)
```

```python
#Criando um histograma através da lista de pontuações em formato int

plt.hist(pontuacoes, density='True', color='purple', edgecolor='black')
plt.title('Histograma de Pontuações')
plt.xlabel('Pontuações')
plt.ylabel('Probabilidade')
plt.grid()
plt.show()
```

```python
#Utilizando um dicionário para converter os numeros em escrita por extenso
#Gerando uma nuvem de palavras com as pontuações
#Gravando a imagem em um arquivo .png

dic_converte = {'1':'um', '2':'dois', '3':'três', '4':'quatro', '5':'cinco', '6':'seis', '7':'sete', '8':'oito', '9':'nove', '10':'dez'}

nova_lista = []
for i in lista_pontuacao:
    nova_lista.append(dic_converte[i])

nova_lista_pronta = ','.join(nova_lista)

print(nova_lista_pronta)

nuvem_palavras = WordCloud(background_color='black', width=800, height=400).generate(nova_lista_pronta)

plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis('off')
plt.show()
nuvem_palavras.to_file('Nuvem de palavras.png')
```

```python
#Análise e conclusão:
#Através da manipulação de arquivos de texto é possível filtrar e limpar os dados que desejamos utilizar.
#O histograma é uma forma de facilitar a visualização dos dados tratados, possibilitando que informações
#possam ser melhor analizadas, principalmente quando trabalhamos com uma grande quantidade de dados.
#A nuvem de palavras é uma forma interessante e divertida de se analisar dados que foram previamente tratados,
#principalmente quando estamos trabalhando com strings ao invés de números inteiros.
```
